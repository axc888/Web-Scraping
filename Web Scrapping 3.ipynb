```python
# import libraries
import urllib2
from bs4 import BeautifulSoup
from pprint import pprint
import re
import pandas as pd
# import sys
# reload(sys)
# # sys.setdefaultencoding('utf8')
```

```python
# specify the url
# https://www.collegesearch.in/bca-colleges-india?page=159
college_name=[]
sr_rating=[]
degree_name=[]
total_duration=[]
total_fees=[]
affliation=[]
for i in range(1,20):
    quote_page = "https://www.collegesearch.in/bca-colleges-india?page="+str(i)
    print(quote_page)
    page = urllib2.urlopen(quote_page)
    # parse the html using beautiful soup and store in variable `soup`
    soup = BeautifulSoup(page, "html.parser")
    name = soup.findAll('div', attrs={'class' : "media-body clg_head"})

# College Name    
    for h in name:
        h2 = h.find('h2')
        a=h2.find('a')
        college_name.append(a.attrs['title'].encode("utf-8"))

    print(len(college_name))
    
# SR Rating
# "Smart Quotient calculated on a gamut of factors that govern quality of an institute."
    srratingall = soup.findAll('div', attrs={'class' : "organic-sr0"})
  
    for srating in srratingall:
        sr_rating.append(srating.attrs['data-percent'].encode("utf-8"))
#     print(len(sr_rating))
    

# Degree Name
    degreeall = soup.findAll('div', attrs={'class' : "col-xs-9 clg_quick_guide"})
     
    for dname in degreeall:
        p = dname.find('p').text.encode("utf-8").strip('\n')
        degree_name.append(p)
#     print(len(degree_name))
    
    restall=soup.findAll('dl', attrs={'class' : "dl-horizontal mg-0"})

    for element in restall:
        dt=element.find_all('dt') 
        dd=element.find_all('dd')
        print("dt",len(dt))
        if dt[0].text.encode("utf-8").find("Duration:")!=-1 :
            total_duration.append(dd[0].text.encode("utf-8")) 
        else:
            total_duration.append("NA")
        if(len(dt)>1): 
            if dt[1].text.encode("utf-8").find("Fees:")!=-1 :
                total_fees.append(dd[1].text.encode("utf-8")) 
            else:
                total_fees.append("NA")
        else:
            total_fees.append("NA")
        if(len(dt)>2):    
            if dt[2].text.encode("utf-8").find("Institute Type:")!=-1 :
                affliation.append(dd[2].text.encode("utf-8")) 
            else:
                affliation.append("NA")
        else:
            affliation.append("None")
            

#     print(len(total_duration))
#     print(len(total_fees))
    print(len(affliation))
    
    
    
percentile_list = pd.DataFrame(
    {'College_Name':college_name,
     'SR_Rating': sr_rating,
     'degree':degree_name,
     'duration': total_duration,
     'Total Fees':total_fees,
     'affliation':affliation
    })

# print(percentile_list)




# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('bca_college_list.xlsx', engine='openpyxl')

# Convert the dataframe to an XlsxWriter Excel object.
percentile_list.to_excel(writer, sheet_name='Sheet1')

# Close the Pandas Excel writer and output the Excel file.
writer.save()

print("Ended")
```
